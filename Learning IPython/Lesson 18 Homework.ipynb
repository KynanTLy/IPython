```python
listOfRows = []
with open(theCSV, 'r') as file: # This makes sure that file is closed after reading
    data = csv.reader(file)
    for row in data:
        listOfRows.append(row)
file.closed
```

```python

categories = listOfRows[0]
```

```python
category = [77, 81]
```

```python
bothList = []
for things in category:
    words = []
    for row in listOfRows[1:]: # We iterate over the rows skipping the header row
        
        
        if row[things] != "":
            words.append(row[0])
            
    bothList.append(words)
        
```

```python
train = [
   ('I attended the hotel in Rio.', 'travel'),
    ('Crusading is fun!.', 'travel'),
    ('Rome conquered many places.', 'travel'),
    ('Did you get admittance at Harvard?', 'travel'),
    ('Can I stay home?', 'stay'),
    ('I do not like to be contained', 'stay'),
    ('I do not want to leave', 'stay'),
    ('Do we have to move to another country?', 'stay')
]
```

```python
test = [
     ('Dancing is the life!', 'travel'),
     ('Watchig tv with my cat', 'stay'),
     ("Do you want to stay at my place?", 'stay'),
     ("How was Rome?", 'travel'),
    ('I wish I was at my country home', 'stay'),
    ("I like to have fun and do not want to leave", 'travel')
]
```

```python
from textblob.classifiers import NaiveBayesClassifier
cl = NaiveBayesClassifier(train)
```

```python
for pair in test:
    item = pair[0]
    blob = TextBlob(item, classifier=cl)
    print(item)
    print(blob.classify())
    print("====")
```

```python
cl.accuracy(test)
```
