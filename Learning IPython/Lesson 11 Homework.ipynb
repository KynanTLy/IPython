```python
import re

#Some HTML TEXT
test = '''<html>
<head>
<title>An example text for regular expressions</title>
</head>
<body>
<p>Here is an example of a using <i>regular expressions.</i> This example
shows how you can do different things with regular expressions. Don't be shy to try them.</p>

<p>Some of the difficulties that we can have are finding and tokenizing hyphenated words 
like tip-off and long-term.</p>

<p>Here are some dates: 1987, 1990, 1389, 2017 and 2027.</p>

<p>Here are some words that are similar woman, women, ...</p>

<p>Prepared 2017.10.15</p>
</body>
</html>'''

#Find all instance of 
matches = re.findall(r'wom.*',test)
print(matches)

print("===========\n")

#Find  all valyes
print(re.findall(r'<.*>',test))

print("===========\n")

#Non greedy regre
print(re.findall(r'<.*?>',test))

#Replace all the tag

print("===========\n")

results = re.sub(r'<.*?>',"",test)
print(results)

#Get everything within a tag
results = re.findall(r'<p>.*?</p>',test)
print(results)

print("===========\n")

#Get all dates that start with 1 or 2 and is 4 long
results = re.findall(r'[12]\d\d\d',test) # [0-9]+ , [0-9][0-9][0-9][0-9]
print(results)

print("===========\n")
#Divide into words
results = re.split(r'\W+',test)
results[:5]
print(results)

print("===========\n")
result = re.findall(r'\w+(?=\s*that)', test)
print(result)
```

```python
test1 = '''<html>
<head>
<title>An example text for regular expressions</title>
</head>
<body>
<p>Here is an example of a using <i>regular expressions.</i> This example
shows how you can do different things with regular expressions. Don't be shy to try them.</p>

<p>Some of the difficulties that we can have are finding and tokenizing hyphenated words 
like tip-off and long-term.</p>

<p>Here are some dates: 1987, 1990, 1389, 2017 and 2027.</p>

<p>Here are some words that are similar woman, women, ...</p>

<p>Prepared 2017.10.15</p>
</body>
</html>'''

#Clear tags and newlines
results = re.sub(r'<.*?>',"",test1)
results = re.sub(r'\n'," ",results)
print(results)

#Words
words = re.findall(r'\b(\w[\w-]*[\w\']*)\b',results)
print(words)


#Find all the name
names = re.findall(r'[A-Z][a-z]+', results)
print(names)

#Find all the year
year = re.findall(r'[0-9]{4}', results)
print(year)

sentence = re.findall(r'[A-Z][a-z]?[-\']?[a-zA-Z]*[\w\s]*[\w]*[.]',results)
print(sentence)
```
