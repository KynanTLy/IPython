```python
import nltk
import matplotlib

#Text Path
textpath = "./Charles Dickens.txt"

#Open file
with open(textpath, "r") as file:
    text = file.read()

    
listofToken = nltk.word_tokenize(text)

theText = nltk.Text(listofToken)

#theText.concordance("Tale", width=80)

#Keep graph in line
%matplotlib inline

theText.dispersion_plot(["Sydney","hero"])

print(theText.count("Sydney"), " ", theText.count("hero"))

#Parse for only words
theText = [token for token in theText if token[0].isalpha()]

#Get the frequency of the text
Freqs = nltk.FreqDist(theText)

#Display the top 25 frequnt word and the amount
Freqs.tabulate(25)

#Plots the top 25 frequency
Freqs.plot(25)

print("=====ADD IN STOP WORDS=====")

stopwords = nltk.corpus.stopwords.words("english")

#Remove the stop words from the list
theText = [token for token in theText if token not in stopwords]

#Find the frequency distribution
NewFreqs = nltk.FreqDist(theText)

#Display the first 25 result in a tab form
NewFreqs.tabulate(25)

#Plot the first 25 results
NewFreqs.plot(25)

#Covert the text without stop words as a nltk object
theText = nltk.Text(listofToken)

#Find top 10 collocation (words next to eachother)
theText.collocations(10)

#Find words used in a similar context to Sydney
theText.similar("Sydney")

theText.common_contexts(["Sydney", "blood"],10)

print("====find all====")

#If I put a () around the <> it will just print the result without the token I was using as a searcher
theText.findall("<Sydney><Carton>(<.*><.*><.*><.*>)")

theText.concordance("Sydney Carton", width=80)
```
